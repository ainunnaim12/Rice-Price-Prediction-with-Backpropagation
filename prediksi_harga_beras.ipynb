```python
import pandas as pd
d = pd.read_csv("hasil_norm.csv")
print(d)
```

```python
X_train= d.iloc[:,0:4].values
X_test = d.iloc[:,-1].values
print(X_train)
print(X_test)
```

```python
import tensorflow as tf #backpro
model = tf.keras.models.Sequential()

model.add(tf.keras.layers.Dense(units= 4, activation='relu')) #node input
model.add(tf.keras.layers.Dense(units= 3, activation='relu')) #node hidden
model.add(tf.keras.layers.Dense(units= 1, activation='sigmoid')) #node output
model.compile(loss='mean_squared_error', optimizer=tf.keras.optimizers.Adam(0.1))

model.fit(X_train, X_test, epochs= 2000, batch_size = 34) #melakukan training


```

```python
hasil_prediksi = model.predict(X_train)
```

```python
print(hasil_prediksi)
```

```python
from sklearn.metrics import mean_squared_error
akurasi = mean_squared_error(X_test,hasil_prediksi)
print(akurasi)
```

```python
import matplotlib.pyplot as plt
a=range(len(X_train))
plt.plot(a,X_test,color='red')
plt.plot(a,hasil_prediksi,color='blue')
plt.title('Hasil Prediksi Harga Rata-Rata Beras Primium')
plt.xlabel('Data Bulan')
plt.ylabel('Data Harga Beras')
plt.show
```
